```python
from re import A
from pydub import AudioSegment
import math
import os

class SplitWavAudio():
    def __init__(self, filepath, output_folder, output_name):
        self.output_folder = output_folder
        self.output_name = output_name
        self.filepath = filepath
        self.audio = AudioSegment.from_wav(self.filepath)
    
    def get_duration(self):
        return self.audio.duration_seconds
    
    def single_split(self, from_sec, to_sec, split_filename):
        t1 = from_sec * 1000
        t2 = to_sec * 1000
        split_audio = self.audio[t1:t2]
        split_audio.export(os.path.join(self.output_folder, split_filename), format="wav")
        
    def multiple_split(self, sec_per_split, total_sec, filecount = 0):
        #total_sec = math.ceil(self.get_duration())
        a = filecount
        for i in range(0, total_sec, sec_per_split):
            a += 1
            split_fn = self.output_name + '_' + str(a) + ".wav"
            self.single_split(i, i + sec_per_split, split_fn)
            if i >= total_sec:
              print('All splited successfully')
```

```python
def find_wav_filenames(path_to_dir, suffix=".wav"):
    filenames = os.listdir(path_to_dir)
    return [os.path.join(path_to_dir, filename) for filename in filenames if filename.endswith(suffix)]

def generate_audio_segments(path_to_dir, output_folder, name, sec_per_split, total_sec):
  a_interval = 0
  filepaths = find_wav_filenames(path_to_dir)
  for path in filepaths:
    split_audio = SplitWavAudio(path,output_folder,name)
    split_audio.multiple_split(sec_per_split=sec_per_split, total_sec = total_sec, filecount = a_interval)
    a_interval = int(total_sec / sec_per_split)
```
